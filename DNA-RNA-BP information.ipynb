```python
test = pd.read_csv("test.csv",header=None,sep='>')
ttest = test[1].dropna(axis=0)
ttest = ttest.reset_index(drop=True)
ttest
```

```python
test2 = pd.read_csv("dnatest.csv",header=None,sep='>')
ttest2 = test2[1].dropna(axis=0)
ttest2 = ttest2.reset_index(drop=True)
ttest2
tts2 = test2[0].dropna(axis=0)
tts2 = tts2.reset_index(drop=True)
tts2
```

```python
DNAsequence = pd.merge(ttest2,tts2,left_index=True, right_index=True,how='left')
DNAsequence = DNAsequence.set_index(1)
DNAsequence
```

```python
DNAsequence['sequence length'] = DNAsequence[0].str.count('')-1
DNAsequence
```

```python
RNAsequence['sequence length'] = RNAsequence[0].str.count('')-1
RNAsequence
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

# DNAsequence와 RNAsequence의 'sequence length' 열에서 데이터 가져오기

# DNA sequence의 히스토그램 그리기 (예시 색상: 파랑)
ax = sns.histplot(data=DNAsequence['sequence length'], color='blue', label='DNA binding Sequence', alpha=0.5)

# RNA sequence의 히스토그램 그리기 (예시 색상: 빨강)
ax = sns.histplot(data=RNAsequence['sequence length'], color='red', label='RNA binding Sequence', alpha=0.5)

# 범례 추가
ax.legend()
ax.set_xlim(0, 5000)
# 그래프 보여주기
plt.show()
```
